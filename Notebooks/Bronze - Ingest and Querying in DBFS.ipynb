# Bronze Layer: Raw Data Ingestion

This notebook demonstrates the ingestion of raw data into the bronze layer of a Lakehouse architecture using Databricks, PySpark, and Delta Lake. The bronze layer stores raw, unprocessed data as Delta tables, serving as the foundation for further data cleansing and transformation.

---

## 1. Schema Creation <a name="schema-creation"></a>

Create the `bronze` schema if it does not already exist.

```python
%sql
CREATE SCHEMA IF NOT EXISTS bronze;
```

## 2. Raw Data Ingestion <a name="raw-data-ingestion"></a>

### Trips Data <a name="trips-data"></a>

Read the raw trips CSV file, infer the schema, and write the data as a Delta table in the bronze layer.

```python
trips_bronze = (
    spark.read
    .format("csv")
    .option("header", "false")
    .option("inferSchema", "true")
    .load("/FileStore/tables/trips.csv")
)

trips_bronze.write.format("delta").mode("overwrite").save("/FileStore/delta/bronze/trips")
```

Register the trips Delta file as a managed table in the bronze schema.

```python
spark.sql("""
CREATE OR REPLACE TABLE bronze.trips
USING DELTA
AS
SELECT *
FROM delta.`/FileStore/delta/bronze/trips`
""")
```

## 3. Data Validation <a name="data-validation"></a>

Preview the ingested trips data from the bronze layer.

```python
%sql
SELECT *
FROM delta.`/FileStore/delta/bronze/trips` Limit 10
```

### Riders Data <a name="riders-data"></a>

Read the raw riders CSV file, infer the schema, and write the data as a Delta table in the bronze layer.

```python
riders_bronze = (
    spark.read
    .format("csv")
    .option("header", "false")
    .option("inferSchema", "true")
    .load("/FileStore/tables/riders.csv")
)

riders_bronze.write \
    .format("delta") \
    .mode("overwrite") \
    .save("/FileStore/delta/bronze/riders")
```

Register the riders Delta file as a managed table in the bronze schema.

```python
spark.sql("""
CREATE OR REPLACE TABLE bronze.riders
USING DELTA
AS
SELECT *
FROM delta.`/FileStore/delta/bronze/riders`
""")
```

Display metadata and schema information for the `bronze.riders` table to validate the ingestion and structure of the raw riders data.

```python
spark.sql("DESCRIBE DETAIL bronze.riders").show(truncate=False)
spark.table("bronze.riders").printSchema()
```
